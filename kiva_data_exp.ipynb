```python
y = dfNoKenya['status']
y = pd.DataFrame(y =="paid")
#X = pd.DataFrame(dfLoans[['lender_count','journal_totals.entries']])
x = pd.DataFrame(dfNoKenya[['terms.loan_amount','partner_id','journal_totals.entries','lender_count','sector']])
#X = pd.DataFrame(dfLoans[['funded_amount','lender_count','journal_totals.entries]])
#print Y

is_train = pd.Series(np.random.uniform(0, 1, len(y.index)) <= .80)

x = pd.concat([x,pd.get_dummies(x['sector']),pd.get_dummies(x['partner_id'])],axis=1)
#x = pd.concat([x,pd.get_dummies(x['partner_id']),pd.get_dummies(x['location.country'])],axis=1)
x= x.drop('partner_id',1)
x= x.drop('sector',1)
#x= x.drop('location.country',1)

x_train, x_test = [], []
y_train, y_test = [], []

for i in range(len(is_train)):
    if is_train[i]==True:
        x_train.append(x.iloc[i]) 
        y_train.append(y.iloc[i])
    else:
        x_test.append(x.iloc[i])
        y_test.append(y.iloc[i])

clf = sklrfc(n_jobs=2)
clf.fit(x_train, y_train)


y_pred = clf.predict_proba(x_test)
#y_pred = clf.predict(x_test)
scores = []
for a in y_pred: scores.extend([a[1]]) #scores.extend(a[1])
scores = np.array(scores)

y_out = []

for i in scores: 
    if i<0.9: 
        y_out.append(False)
    else:
        y_out.append(True)

fpr, tpr, thresholds = sklm.roc_curve(y_test,scores)

#pd.crosstab(np.ravel(y_test), np.ravel(y_pred), rownames=['actual'], colnames=['preds']) 
pd.crosstab(np.ravel(y_test), np.ravel(y_out), rownames=['actual'], colnames=['preds']) 
#return dfLoan

```

```python
#for a in clf.predict_proba(x_test).T[0]: print a
#np.array([0.1, 0.4, 0.35, 0.8])
plt.plot(fpr,tpr)
plt.plot(range(0,2),range(0,2))
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
```
